```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random waypoints
num_waypoints = 30
x = np.random.randint(0, 80, size=num_waypoints)
y = np.random.randint(0, 80, size=num_waypoints)

# Plot the waypoints
plt.scatter(x, y, color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Waypoints')
plt.grid(True)
plt.show()

# Create the route from the waypoints
route = np.column_stack((x, y))

# Print the route coordinates
print("Route coordinates:")
print(route)
```
