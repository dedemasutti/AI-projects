```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Carregar a imagem para o ambiente do Google Colab
uploaded = files.upload()

# Ler a imagem com OpenCV
image = cv2.imread(next(iter(uploaded)))

# Converter a imagem de BGR para RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Exibir a imagem carregada com matplotlib em alternativa caso dê erro com a função imshow() do openCV.
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python

# Aplicar um filtro de desfoque (blur) na imagem
blurred_image = cv2.GaussianBlur(image, (7, 7), 0)

# Converter a imagem de BGR para RGB
blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

# Exibir a imagem com filtro de desfoque
plt.imshow(blurred_image_rgb)
plt.axis('off')
plt.show()
```
