```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Função para identificar pontos de interesse em uma imagem
def identificar_pontos_interesse(image_path):
    # Carregando a imagem
    image = cv2.imread(image_path)
    if image is None:
        print("Imagem não encontrada!")
        return

    # Convertendo a imagem para escala de cinza
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Aplicando um filtro Gaussiano para reduzir o ruído
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

    # Detecção de bordas usando o operador Canny
    edges = cv2.Canny(blurred_image, threshold1=30, threshold2=100)

    # Encontrando contornos na imagem
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Desenhando os contornos na imagem original
    image_with_contours = image.copy()
    cv2.drawContours(image_with_contours, contours, -1, (0, 255, 0), 2)

    # Exibindo a imagem com os contornos identificados
    cv2_imshow(image_with_contours)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    # Substitua o caminho pela sua própria imagem capturada pelo drone
    image_path = "/content/area verde.jpg"
    identificar_pontos_interesse(image_path)
```
